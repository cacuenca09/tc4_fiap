# 📈 LSTM Stock Price Predictor — Análise Exploratória

**Empresa:** Apple Inc. (AAPL)  
**Período:** 2018-01-01 → 2024-07-20


```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings('ignore')

SYMBOL     = 'AAPL'
START_DATE = '2018-01-01'
END_DATE   = '2024-07-20'

df = yf.download(SYMBOL, start=START_DATE, end=END_DATE, auto_adjust=True)
print(f'Shape: {df.shape}')
df.head()
```

```python
# ─── Estatísticas descritivas ────────────────────────────────────────────────
df.describe().round(2)
```

```python
# ─── Série temporal do preço de fechamento ───────────────────────────────────
fig, axes = plt.subplots(3, 1, figsize=(14, 12))

axes[0].plot(df.index, df['Close'], color='#2196F3', linewidth=1)
axes[0].set_title(f'{SYMBOL} — Preço de Fechamento Ajustado', fontsize=13)
axes[0].set_ylabel('USD')
axes[0].grid(alpha=0.3)

axes[1].bar(df.index, df['Volume'], color='#90CAF9', alpha=0.7)
axes[1].set_title('Volume de Negociação', fontsize=13)
axes[1].set_ylabel('Ações')
axes[1].grid(alpha=0.3)

# Retornos diários
df['Returns'] = df['Close'].pct_change() * 100
axes[2].plot(df.index, df['Returns'], color='#4CAF50', linewidth=0.8)
axes[2].axhline(0, color='red', linewidth=0.8, linestyle='--')
axes[2].set_title('Retorno Diário (%)', fontsize=13)
axes[2].set_ylabel('%')
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ─── Médias Móveis ───────────────────────────────────────────────────────────
df['MA_20']  = df['Close'].rolling(20).mean()
df['MA_50']  = df['Close'].rolling(50).mean()
df['MA_200'] = df['Close'].rolling(200).mean()

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['Close'],  label='Fechamento', linewidth=1, alpha=0.9)
plt.plot(df.index, df['MA_20'],  label='MA 20',      linewidth=1.2)
plt.plot(df.index, df['MA_50'],  label='MA 50',      linewidth=1.2)
plt.plot(df.index, df['MA_200'], label='MA 200',     linewidth=1.5, linestyle='--')
plt.title(f'{SYMBOL} — Médias Móveis', fontsize=13)
plt.ylabel('USD')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ─── Treinar modelo ──────────────────────────────────────────────────────────
# Execute a célula abaixo para chamar o script de treinamento
import subprocess, sys
result = subprocess.run(
    [sys.executable, '../model/train.py'],
    capture_output=True, text=True
)
print(result.stdout)
if result.returncode != 0:
    print('STDERR:', result.stderr)
```
